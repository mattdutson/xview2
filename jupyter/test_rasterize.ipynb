```python
import os

import numpy as np
import matplotlib.pyplot as plt
import skimage.io
```

```python
plt.rcParams["figure.figsize"] = 12, 12
```

```python
def show(name, save_name=None):
    image = skimage.io.imread(os.path.join("dataset", "train", "images", name) + ".png")
    mask = np.load(os.path.join("dataset", "train", "raster_labels", name) + ".npy")    
    plt.imshow(image)
    buildings = np.ma.masked_where(mask == 0, mask)
    plt.imshow(buildings, cmap="jet", alpha=0.5, vmin=1, vmax=4)
    plt.axis("off")
    if save_name is not None:
        plt.savefig(save_name)
    plt.show()
```

```python
show("santa-rosa-wildfire_00000014_pre_disaster", os.path.join("jupyter", "sample_rasterized.pdf"))
```
